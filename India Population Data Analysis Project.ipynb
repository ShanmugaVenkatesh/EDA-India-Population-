## DATA ANALYSIS

### CORRELATION MATRIX

```python
data_corr = population.corr(method = "pearson")
data_corr
```

## CORRELATION HEATMAP BETWEEN VARIABLES

```python
plt.style.use("dark_background")
plt.figure(figsize=(16,10))
heatmap=sns.heatmap(data_corr,annot=True,square=True ,center=0,vmax=1,vmin=-1,fmt=".2g",linecolor="Black",cmap="viridis",linewidth=1)
heatmap.set_title("Correlation Heatmap between Variable");
```

### ObSERVATION
- Negative correlation between each pairs are follows:<br>
        1. (Year, Migrants) - When days pass, year changes but count of migrants gradually fall off. People reside in a                    place of their own comfort zone. This led to only increase in population, without any replacement in                        population.
        2. (India Global Rank, Year) - As year passes, India is becoming the most populated country in the world, this                    represents the highest in the Global Rank.
        3. (Migrants , Population) - India's Population is escalating but the migrants gradually diminishing.
        4. (Population , India Global Rank) - As the Population grows, India is becoming the top populated country in the                  of world.
- Positive correlation between each Pairs are follows: <br>
        1. (Population, Density) - When Population grows, population per km square also rises.
        2. (Median Age, Urban pop%) - As Urban pop% rises, that leads to rise in Median Age
        3. (Population, World Population) - As India's Population escalates, then obviously, it constitutes huge portion 
        of the world population. This permits  surge the World Population.
        4. (Year, Urban population) - As year passes, the population in the urban areas in India increases.

## COVARIANCE HEATMAP

```python
plt.figure(figsize=(16,10))

heatmap=sns.heatmap(population.cov(),annot=True,center=0,vmax=1,vmin=-1,fmt=".2g",linecolor="Black",cmap="icefire",linewidth=1)
heatmap.set_title("Covariance Heatmap between Variable");
```

### OBSERVATION
- Negative value of covariance between Migrants and Population indicates that the variables are inversely related.Such inversely related pairs of variables are,
    - (Migrants, Population)
    - (Year/Population, Year % Change) 
    - (Yearly Change, Year/Population)
    - (India Global Rank, Year/Population/Migrants)
    - (Fertility rate, Population/Year)
    - (Median Age, India Global Rank)

## POPULATION vs MIGRANTS

```python
plt.figure(figsize=(14,7))
sns.regplot(data=population,x="Population",y="Migrants (net)",color="c").set_title("Population vs Migrants")
```

## OBSERVATION
- As years pass from 1955 to 2022, very less count of emigration had taken place throughout the world. 
- The growth of the population varies inversely proportional to the Migrants.
- The Regression Curve, it is explicit that the migrants count retards and population growth escalates.

## RELATION BETWEEN URBAN POPULATION AND INDIA'S POPULATION

## RELATION BETWEEN INDIA'S POPULATION AND WORLD POPULATION

```python
plt.style.use("dark_background")
plt.figure(figsize=(8,5))
sns.kdeplot(population["World Population"],color='blue',label='World Population',shade=True)
sns.kdeplot(population["Population"],color='red',label="India's Population",shade=True)
plt.legend()
```

###  Observation from density plot between India's Population and World Population:
- As the peak of India's Population occurs way too higher than the peak of world population Density. Hence, India's Population escaltes in higher frequency than that of the World's Population.
- This may lead to many consequences, that are faced by China because of overpopulation.

## URBAN POPULATION CONTRIBUTION IN THE COUNTRY'S POPULATION

```python
population.set_index('Year', inplace=True)
population[['Urban Population','Population']].head(18).plot.bar(stacked=True, color = ['darkviolet','deeppink'],figsize = (12,8))
plt.title('Urban population VS country population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

## OBSERVATION
- Urban Populatimon contribution towards country's population plays a vital role in rise of India's Population

```python
plt.figure(figsize=(12,6))
future_predict = population[population['Year'] > 2022]
plt.bar(future_predict['Year'],future_predict['Population'], color ='yellow', width = 2)
plt.xlabel("Year")
plt.ylabel("Population  (billion)")
plt.title("Future Prediction of Indian Population upto 2050")
```

## OBSERVATION
- India's Population rapidly escalating from 2025 to 2050.
- It is going to reach about 1.66 billion in the year of 2050.
- In 2045, the Population is about - 1.6 billion
- In 2040, the Population is about - 1.5 billion
- In 2025, the Population is about - 1.40 billon
